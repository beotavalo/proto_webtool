```python
import pandas as pd
#Read individuals risk file
df = pd.read_excel('/content/PCA_georisk.xlsx')
df.head()
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Aplicar el algoritmo de PCA
pca = PCA(n_components=1,svd_solver='randomized')  # Establecer el número de componentes
result =pca.fit_transform(df[['R_Volc','R_Sism','R_Deslav','R_Inund']])
result
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Seleccionar solo las columnas específicas para el PCA
columnas_pca = df[["R_Volc", "R_Sism", "R_Deslav", "R_Inund"]]

# Estandarizar los datos
scaler = StandardScaler()
columnas_pca_estandarizadas = scaler.fit_transform(columnas_pca)

# Aplicar PCA con el número de componentes adecuado
pca = PCA(n_components=1)
resultados_pca = pca.fit_transform(columnas_pca_estandarizadas)

# Visualizar los componentes principales
print(resultados_pca)
```

```python
df['georisk'] = 1- ((resultados_pca - min(resultados_pca))/(max(resultados_pca)-min(resultados_pca)))
df
```
